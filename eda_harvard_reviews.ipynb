```python
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/harvard_rewiews/harvard_reviews.csv')

```

```python
for column in data.columns:
    if data[column].dtype == 'object':
        # Jika kolom bertipe objek (teks), isi dengan modus (nilai yang paling sering muncul)
        mode_value = data[column].mode()[0]
        data[column] = data[column].fillna(mode_value)
    else:
        # Jika kolom numerik, isi dengan nilai rata-rata
        mean_value = data[column].mean()
        data[column] = data[column].fillna(mean_value)
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
# Mengecek apakah ada duplicate di seluruh kolom
check_duplicate = data.duplicated().sum()

print(f"Jumlah data yang duplikat = {check_duplicate}")
```

```python
# Handling duplicate
data = data.drop_duplicates()

#tapi disni data saya sudah tidak ada data yang duplikat
```

```python
# Mengecek duplicate setelah di-handle
handle_duplicate = data.duplicated().sum()

print(f"Jumlah data yang duplikat = {handle_duplicate}")
```
